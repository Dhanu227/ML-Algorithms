```python
import pandas as pd
import numpy as np

# Initialize weights DataFrame with 9 features and 3 weight vectors (columns)
w = pd.DataFrame(data=[
    [0.4, 0.2, 0.5],
    [0.9, 0.5, 0.3],
    [0.2, 0.3, 0.9],
    [0.6, 0.4, 0.8],
    [0.9, 0.8, 0.5],
    [0.4, 0.7, 0.2],
    [0.3, 0.6, 0.4],
    [0.5, 0.6, 0.2],
    [0.7, 0.2, 0.7]
], columns=['w1', 'w2', 'w3'])
```

```python
# Initialize the animal feature DataFrame
df = pd.DataFrame(data={
    "Animal": ["Dove", "Owl", "Dog", "Lion", "Horse"],
    "is small": [1, 1, 0, 0, 0],
    "is medium": [0, 0, 1, 0, 0],
    "is big": [0, 0, 0, 1, 1],
    "2 legs": [1, 1, 0, 0, 0],
    "4 legs": [0, 0, 1, 1, 1],
    "feather": [1, 1, 0, 0, 0],
    "hunt": [0, 1, 0, 1, 0],
    "run": [0, 0, 1, 1, 1],
    "fly": [1, 1, 0, 0, 0]
})

print("Animal Features DataFrame:\n", df)
```

```python
# Initialize output DataFrame
op = pd.DataFrame([0, 0, 0, 0, 0], columns=["Output"])
arr = np.array(["dove", "owl", "dogs", "lion", "horse"])  # Array of animal names
lr = 1.2  # Initial learning rate
temp = np.array([0, 0, 0], dtype=float)  # Array to store distances

# Loop for 10 epochs
for k in range(0, 10):
    lr = lr / 2  # Halve the learning rate for each epoch
    print(f"Epoch: {k + 1}")
    print(f"Learning Rate: {lr}")

    # Loop over all entries in arr
    for j in range(0, 5):
        print("For:", arr[j])

        # Extract features from the DataFrame and convert to float
        x = df.iloc[j, 1:].astype(float)
        print("Feature Vector:", x.values)

        # Calculate distances from the current weight vectors
        for i in range(0, 3):
            y = w.iloc[:, i].values  # Ensure 'y' is a column vector matching 'x' length
            temp[i] = np.sum((x.values - y) ** 2)

        print("D1, D2, D3 are", temp)

        # Find the minimum distance and corresponding weight vector index
        id = np.argmin(temp)
        print(f"Minimum of them is {temp[id]}, closest to W{id + 1}")

        # Update the output with the selected weight vector
        op.iloc[j] = "w" + str(id + 1)
        print("Changing Weights for W", id + 1)
        print("Old Weights:", w.iloc[:, id])

        # Update the weights for the selected vector
        v = df.iloc[j, 1:].values.astype(float)
        z = w.iloc[:, id].values
        w.iloc[:, id] = z + lr * (v - z)

        print("New Weights Updated:")
        print(w.iloc[:, id])

    # Display updated weights after each epoch
    print("Updated Weights after Epoch", k + 1)
    print(w.round(3))
```
